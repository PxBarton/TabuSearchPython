```python
sheet = book.sheet_by_index(3)
kRows = 30
dRows = 30
C = np.array(sheet.row_values(1, 1, 101), dtype=int)
C = list(C)

if kRows > 0:
    kConstraints = []
    
    for i in (range(kRows)):
        Ak = np.array(sheet.row_values(35 + i, 1, 101), dtype=int)
        Ak = list(Ak)
        kConstraints.append(Ak)
        

else:
    kConstraints = []


if dRows > 0:
    dConstraints = []
    
    for i in (range(dRows)):
        Ad = np.array(sheet.row_values(5 + i, 1, 101), dtype=int)
        Ad = list(Ad)
        dConstraints.append(Ad)        
        
else:
    dConstraints = []
    
Bk = np.array(sheet.col_values(103, 35, 35 + (kRows)), dtype=int)
Bd = np.array(sheet.col_values(103, 5, 5 + (dRows)), dtype=int)

kConstraints
```
