```python
import numpy as np
import math
import xlrd
import itertools as it
import csv
import pandas as pd
import scipy.stats as stat
```

```python
book = xlrd.open_workbook('case3and6 binary feasible solutions for difficult MDMKPs 02 03 2022 BIP.xlsx')
book.sheet_names()
```

```python
sheet = book.sheet_by_index(5)
kRows = 30
dRows = 30
C2 = np.array(sheet.row_values(1, 1, 101), dtype=int)
C2 = list(C2)

if kRows > 0:
    kConstraints = []
    Ak2 = np.array(sheet.row_values(38, 1, 101), dtype=int)
    Ak2 = list(Ak2)
    Ak2.append("\\n")
    kConstraints.append(Ak2)
    for i in (range(kRows-1)):
        Ak2_next = np.array(sheet.row_values(38 + (i + 1), 1, 101), dtype=int)
        Ak2_next = list(Ak2_next)
        Ak2_next.append("\\n")
        kConstraints.append(Ak2_next)
        

else:
    Ak2 = np.array([])


if dRows > 0:
    dConstraints = []
    Ad2 = np.array(sheet.row_values(8, 1, 101), dtype=int)
    Ad2 = list(Ad2)
    Ad2.append("\\n")
    dConstraints.append(Ad2)
    for i in (range(dRows-1)):
        Ad2_next = np.array(sheet.row_values(8 + (i + 1), 1, 101), dtype=int)
        Ad2_next = list(Ad2_next)
        Ad2_next.append("\\n")
        dConstraints.append(Ad2_next)        
        
else:
    Ad2 = np.array([])
    
Bk2 = np.array(sheet.col_values(103, 38, 38 + (kRows)), dtype=int)
Bd2 = np.array(sheet.col_values(103, 8, 8 + (dRows)), dtype=int)
```

```python
with open('ObjCoeffs2.csv', mode='w') as ObjCoeffs:
    probWrite = csv.writer(ObjCoeffs, delimiter = ' ')
    probWrite.writerow(C2)
```

```python
with open('LHSknapsack2.csv', mode='w') as LHSCoeff:
    probWrite = csv.writer(LHSCoeff, delimiter = ' ')
    for row in range(len(kConstraints)):
        probWrite.writerow(kConstraints[row])
```

```python
with open('LHSdemand2.csv', mode='w') as LHSCoeff:
    probWrite = csv.writer(LHSCoeff, delimiter = ' ')
    for row in range(len(dConstraints)):
        probWrite.writerow(dConstraints[row])
```

```python
with open('RHSknapsack2.csv', mode='w') as RHSCoeff:
    probWrite = csv.writer(RHSCoeff, delimiter = ' ')
    probWrite.writerow(Bk2)
```

```python
with open('RHSdemand2.csv', mode='w') as RHSCoeff:
    probWrite = csv.writer(RHSCoeff, delimiter = ' ')
    probWrite.writerow(Bd2)
```
